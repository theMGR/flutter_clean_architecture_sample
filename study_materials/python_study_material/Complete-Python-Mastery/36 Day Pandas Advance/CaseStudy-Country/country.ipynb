```python
Country = []
Capital = []
Official_language = []
with open('world.csv', 'r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    for i in csv_reader:
        #print(i)
        # access single column 
        Country0 = i[0]
        Capital8 = i[8]
        Official_language23 = i[24]
        #print(col1)
        Country.append(Country0)
        Capital.append(Capital8)
        Official_language.append(Official_language23)

#print(Country)
#print(Capital)
#print(Official_language)

size = len(Country)

for i in range(size):
  print(f'{Country[i]} | {Capital[i]} | {Official_language[i]}')
```

```python
data = {
    'Country': Country,
    'Capital': Capital,
    'Official_language':Official_language
}

df = pd.DataFrame(data)
df = df.iloc[1:]
# print(df)
df.to_csv('bountry.csv', index=False)
```

```python
df = pd.DataFrame(data)
df = df.iloc[1:10]
# print(df)
df.to_csv('bountry10.csv', index=False)
```
