```python
import pandas as pd
djia_data = pd.read_csv('HistoricalPrices.csv')
djia_data.head()
```

```python
# show table structure
djia_data.info()
```

```python
djia_data = djia_data.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
```

```python
djia_data['Date'] = pd.to_datetime(djia_data['Date'])
djia_data = djia_data.sort_values(by = 'Date')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Line Plots with a Single Line
plt.plot(djia_data['Date'], djia_data['Close'])
plt.show()
```

```python
# Line Plots with Multiple Lines
plt.plot(djia_data['Date'], djia_data['Open'])
plt.plot(djia_data['Date'], djia_data['Close'])
plt.plot(djia_data['Date'], djia_data['Low'])
plt.plot(djia_data['Date'], djia_data['High'])
plt.show()
```

```python
plt.plot(djia_data['Date'], djia_data['Open'], label = 'Open')
plt.plot(djia_data['Date'], djia_data['Close'], label = 'Close')
plt.legend()
plt.show()
```

```python
# Drawing Bar Plots
plt.bar(djia_data['Date'], djia_data['Close'])
plt.show()
```

```python
# Drawing Bar Plots month wise
djia_data['Month'] = djia_data['Date'].dt.month
monthly_averages = djia_data.groupby('Month').mean()
plt.bar(monthly_averages.index, monthly_averages['Close'])
plt.show()
```

```python
# Import the calendar package 
from calendar import month_name

# Order by months by chronological order
djia_data['Month'] = pd.Categorical(djia_data['Date'].dt.month_name(), month_name[:])

# Group metrics by monthly averages
djia_monthly_mean = djia_data \
    .groupby('Month') \
    .mean() \
    .reset_index()

djia_monthly_mean.head(6)
#print(djia_monthly_mean)
```

```python
plt.bar(djia_monthly_mean['Month'], height = djia_monthly_mean['Close'])
plt.show()
```

```python
# Horizontal Bar Plots

plt.barh(djia_monthly_mean['Month'], width = djia_monthly_mean['Close'])
plt.show()
```

```python
# Drawing Scatter Plots
plt.scatter(djia_data['Open'], djia_data['Close'])
plt.show()
```

```python
# Scatter Plots with a Trend Line

plt.scatter(djia_data['Open'], djia_data['Close'])
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('Open Price vs Close Price')
plt.show()
```

```python
z = np.polyfit(djia_data['Open'], djia_data['Close'], 1)
p = np.poly1d(z)


plt.scatter(djia_data['Open'], djia_data['Close'])
plt.plot(djia_data['Open'], p(djia_data['Open']))
plt.show()
```

```python
# Setting the Plot Title and Axis Labels
plt.scatter(djia_data['Open'], djia_data['Close'])
plt.show()
```

```python
plt.plot(djia_data['Date'], djia_data['Open'], color = 'black')
plt.plot(djia_data['Date'], djia_data['Close'], color = 'red')
plt.show()
```

```python
# change the color of bars   

plt.bar(djia_monthly_mean['Month'], height = djia_monthly_mean['Close'], color = 'red', edgecolor = 'black')
plt.show()
```

```python
# Sort the dataframe by 'Month'
djia_monthly_mean_srtd = djia_monthly_mean.sort_values(by='Month')

# change the fill color of bars
plt.bar(djia_monthly_mean_srtd['Month'], height = djia_monthly_mean_srtd['Close'], color = ['blue', 'green', 'yellow', 'pink', 'gray', 'red', 'purple', 'orange', 'brown', 'cyan', 'magenta', 'black'])
plt.show()
```
